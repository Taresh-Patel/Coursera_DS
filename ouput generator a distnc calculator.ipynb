```python
import pandas as pd
import os
import urllib.request
import json
import requests
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
import pgeocode

os.chdir(r'G:\Shared drives\Business Analytics\CRM\Selections\New store opeining Fort William')
input_file_name = 'CustomerPostCodes.csv'
```

```python
Postcodes=pd.read_csv(input_file_name)
```

```python
Postcodes['Postcode_Cleaned'] = Postcodes['Postcode'].str.replace(' ', '').str.upper() 
Postcodes['Postcode'] = Postcodes['Postcode'].str.upper() 
```

```python
dist = pgeocode.GeoDistance('GB')
nomi = pgeocode.Nominatim('GB')
```

```python
Postcodes[['latitude',	'longitude']] = nomi.query_postal_code(list(Postcodes['Postcode']))[['latitude',	'longitude']]
```

```python
dest = ['PH33 6RU']*len(list(Postcodes['Postcode']))
```

```python
Postcodes['Distance_KM'] = dist.query_postal_code(list(Postcodes['Postcode']), dest)
```

```python
Postcodes_filtered = Postcodes.copy()[Postcodes['Distance_KM'] <= 25 ]
```

```python
Postcodes_filteredNAN = Postcodes.copy()[Postcodes['Distance_KM'].isnull()]
```
